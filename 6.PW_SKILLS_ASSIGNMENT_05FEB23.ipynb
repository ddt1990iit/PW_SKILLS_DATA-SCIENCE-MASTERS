## Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.


Ans:

**CLASS:**: A class is the blueprint of the object.It is also referred as the object constructor.It is a user dfined data-structore that binds the data menmbers and methods into the single units. As an example we can consider **Car** as the class and **toyota**,**Innova**,**maruti** as the object of the Car Class. 

**Object**: Object: An object is an instance of a class. It is a collection of attributes and methods. We use the object of a class to perform actions. Objects of the class has certain characteristics :
* Identity
* State
* Behavior

When an object of a class is created, the class is said to be instantiated. All the instances share the attributes and the behavior of the class. But the values of those attributes, i.e. the state are unique for each object. A single class may have any number of instances.

```python
# Expmle code to define Class and objects
class Car:
    # Methods of the Car Class or Initialization method
    def __init__(self,model_name,year):
        self.model_name = model_name ## Attributes are initialized
        self.year = year 
        
    # Customized method to perform operations
        
    def display_att(self):
        print('The model of the car is : {}'.format(self.model_name))
        print('The year of make is : {}'.format(self.year))
        
car1 = Car('Innova',1975) # Instance of Car class or Object
car2 = Car('Toyota',1980) # # Instance of Car class or Object
# access method inside class
car1.display_att()
car1.display_att()
        
```

# Q3. Explain why the __init__() function is used. Give a suitable example.


Ans : **init()** method is referred as the constructor of the class.Constructor is a special method used to create and initialize an object of a class.When a class defines an __init__() method, class instantiation automatically invokes __init__() for the newly created class instance

```python
# Example to demonstrate __init__() method
class bike:
    def __init__(self,bike_price,rate):
        self.bike_price = bike_price
        self.rate = rate
    
    def tax_cal(self):
        road_tax = self.bike_price*self.rate/100
        return road_tax

bike1 = bike(125000,15) # __init__ method of Class bike will be invoked as soon as the instance is created.
# access method inside class
bike1.tax_cal()
```

## Q4. Why self is used in OOPs?


Ans : Whenever we define instance methods for a class, we use **self** as the first parameter. Using self, we can access the instance variable and instance method of the object.

The first argument **self** refers to the current object.

Whenever we call an instance method through an object, the Python compiler implicitly passes object reference as the first argument commonly known as **self**.

It is not mandatory to name the first parameter as a **self**. We can give any name whatever we like, but it has to be the first parameter of an instance method.



## Q5. What is inheritance? Give an example for each type of inheritance.

Ans :
**Inheritance** is the ability to create a new class (child class) from an existing one (parent class). The child class typically inherits the attributes (members and methods) of the parent class, although it can also redefine them.

### **Types of Inheritance:**


# Example 1 : Single Inheritance


```python
# Code to Explain the single inheritance
```

```python
# Parent Class
class A:
    def func1():
        print('This is from the Parent Class A')

# Child Class
class B(A): # Sngle Inhetitance
    def func2():
        print('This is from the Child Class B')
        
child_B = B() # Object of Child Class
B.func2()
B.func1()
```

# Example 2 : Multiple Inheritance

```python
# Code to Explain the Multiple inheritance
```

```python
# Parent Class
class A:
    def func1():
        print('This is from the Parent Class A')
        

class B():
    def func2():
        print('This is from the Parent Class B')
        
# Child Class
class C(A,B): # Multiple inheritance
    def func3():
        print('This is from the Child Class C')
        
child_c = C() # Object of Child Class
C.func1()
C.func2()
C.func3()
```

# Example 3 : Multi-Level Inheritance 

```python
# Parent Class
class A:
    def func1():
        print('This is from the Parent Class A')
        
# Intermediatory Class
class B(A):
    def func2():
        print('This is from the Intermediatory Class B')
        
# Derived Class
class C(B): # Multi-Level inheritance
    def func3():
        print('This is from the Derived Class C')
        
derived_c = C() # Object of Derived Class
C.func1()
C.func2()
C.func3()
```

# Example 4 : Hierarichal Inheritance

```python
# Base class
class Parent_A:
    def func1(self):
        print("This function is in parent class.")
        
# Derived Class 1
 
class Child_B(Parent_A):
    def func2(self):
        print("This function is in child 1.")
        
# Derived Class 2 

class Child_C(Parent_A):
    def func3(self):
        print("This function is in child 2.")

# Derived Class 3
        
class Child_D(Parent_A):
    def func4(self):
        print("This function is in child 3.")
 
 
# Driver's code
object1 = Child_B()
object2 = Child_C()
object3 = Child_D()
object1.func1()
object2.func1()
object1.func2()
object3.func1()
object3.func4()
```

# Example 4 : Hybrid Inheritance


```python
# Base class
class Parent_G:
    def func1(self):
        print("This function is in parent class G")

# Base class
class Parent_F:
    def func2(self):
        print("This function is in parent class F")

# Child Class
class Child_E(Parent_F,Parent_G): # Multiple Inheritance
    def func3(self):
        print("This function is in parent class E")
        
# Derived Class 1
 
class Child_B(Parent_F):
    def func4(self):
        print("This function is in child B.")
        
# Derived Class 2 

# Hierarichal Inheritance

class Child_A(Child_B): 
    def func5(self):
        print("This function is in child A.")

# Derived Class 3
        
class Child_C(Child_B):
    def func6(self):
        print("This function is in child C.")
        
object1 = Child_C()
object2 = Child_A()
object1.func2()
object1.func4()
object1.func6()
```
