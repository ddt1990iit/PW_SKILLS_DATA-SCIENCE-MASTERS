### Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The function should print the sum to the console. For example, if the 'Values' column of df contains the values [10, 20, 30, 40, 50], your function should calculate and print the sum of the first three values, which is 60.



```python
import pandas as pd

df = pd.DataFrame({'values':[10,20,30,40,50]})

def sum_values(df):
    sum = df['values'][0:3].sum()  # Slicing the first three values of df column
    return sum
    
sum = sum_values(df)

print('The data Frame is\n',df)

print(f'The sum of the first three value of the column of Data Frame is :  {sum}' )
```

### Q4. Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column 'Word_Count' that contains the number of words in each row of the 'Text' column.


```python
df = pd.DataFrame({'Text':['I love Data science masters','web development is good course to learn','pwskills is great platform to learn data science']})

# Function to calculate the the number of words in a string

def Word_Count(df):
    for i in range(0,df.size):
        df['Word_count'] = df['Text'].apply(lambda x: len(x.split()))
    return df


df_updated = Word_Count(df)

df_updated
```

## Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email addresses in the format 'username@domain.com'. Write a Python function that creates a new column 'Username' in df that contains only the username part of each email address.

## The username is the part of the email address that appears before the '@' symbol. For example, if the email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your function should extract the username from each email address and store it in the new 'Username' column.



```python
df = pd.DataFrame({'Email':['dharmendra.dubey@gmail.com','ankit.srivastava@gmail.com','sudhanshu@gmail.com']})


def username(df):
    for i in range(0,df.size):
        df['Username'] = df['Email'].apply(lambda x: x.split('@')[0]) # Lambda function to extract the username using split function
    return df

df_updated = username(df)

df_updated
```

## Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean,median, and standard deviation of the values in the 'Values' column.


```python
import pandas as pd

df = pd.DataFrame({'values':[10,20,30,40,50]})

df
```

```python
def mean_values(df):
    mean = df['values'].mean()  # Calculating the mean of the column Values
    return mean 

def median_values(df):
    median = df['values'].median()  # Calculating the median of the column Values
    return median 

def SD_values(df):
    SD = df['values'].std()  # Calculating the standard deviation of the column Values
    return SD 
    

print(f'The mean of values column of Data Frame is :  {mean_values(df)}' )
print(f'The median of values column of Data Frame is :  {median_values(df)}' )
print(f'The Standard Deviation of values column of Data Frame is :  {SD_values(df)}' )
```

## Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days for each row in the DataFrame. The moving average should be calculated using a window of size 7 and should include the current day.


```python
import pandas as pd

df = pd.read_csv('sales_data_sample.csv')

df
```

```python
# calculating simple moving average of sales column(Simple moving Average SMA7)
# using .rolling(window).mean() ,
# with window size = 7

def moving_average(df):
    df['Moving Average'] = df['SALES'].rolling(7,min_periods=1).mean()
    return df

df_updated = moving_average(df)

df_updated
```

```python
df = pd.DataFrame({'Date':['2023-01-01','2023-01-02','2023-01-03','2023-01-04','2023-01-05']})

df
```

```python
# Function to get the weekdays name

def get_weekday(df):
    df['Date'] = pd.to_datetime(df['Date']) # Convert the date into datetime data type
    df['Weekdays'] = df['Date'].dt.day_name()
    return df

df_updated = get_weekday(df)

df_updated   
```

## Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python function to select all rows where the date is between '2023-01-01' and '2023-01-31'.


```python
import pandas as pd

data = {'Date':pd.date_range(start='1/1/2022', end='1/31/2023')}

df = pd.DataFrame(data)

df
```

```python
def get_date(df):
    
    df = df[df['Date'].between("2023-01-01", "2023-01-31")]
    
    return df

df_updated = get_date(df)

df_updated
```
