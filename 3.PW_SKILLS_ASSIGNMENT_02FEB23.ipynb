Ans : Two methods that are available with Tuple data structure are :

* 1. count()
* 2. index()

```python
# Example 1: count()
ex_tuple = (1,2,3,4,2,4,8,9) 
count = ex_tuple.count(2) # method to find the no of occurence '2' in tuple 
print(count)
```

```python
# Exapmle 2 : index()
ex_tuple1 = ('Apple',22,11,'Orange','Apple',11)
id = ex_tuple1.index('Apple') # Method to find the first instance of the element in tuple and return the index of the element
print(id)
```

```python
# Exapmle 3 : index()
ex_tuple1 = ('Apple',22,11,'Orange','Apple',11)
id = ex_tuple1.index('Apple',1) # '1' is the start value of index after which search start
print(id)
```

Ans : **union() :** This the in-built fuction used in collection data types like Set to concatenate only. It will only give the instance of the operation and doesn't change the set on which union() is operated. This method will return a set containing the union of sets. Duplicates are not shown.

```python
# Example : Set
set1 = {"Dharmendra","Dubey"}
set2 = {'Dubey'}
set1.union(set2)
```

```python
print(set1) # Set 1 is unchanged, InPlace= False
```

```python
print(set2) # Set 2 is unchanged, Inplace = False
```

**update() :** This the in-built fuction used to update the set with another set, or any other iterable.Duplicates are not allowed and if item is present in one set only one value will be updated in the set.

```python
# Example : 
set1 = {1,2,3,4,6}
set2 ={1,3,5,7,8}
print("Set1 before update()",set1)
print("\nSet2 before update()",set2)
set1.update(set2) 
print("\nSet1 after update()",set1) # Set 1 will be changed i.e Inplace = True
print("\nSet2 after update()",set2)
```

Ans : Three View objects in dictionaries are following:
* 1. Keys
* 2. Values
* 3. Items

Keys holds all the keys of the dictionary, the values objects holds all the values, whereas the items holds the key-value pairs.

**Three inbult methods in python to display these three objects are following**

* keys(): To extract all the keys in dictionay
* values() : To extract all the value in the dictionary
* items() : To extract all the key-value pair in the dictionary

```python
# Three inbuilt methods to display three view object in the given dictionary

dict1 = {'Sport': 'Cricket' , 'Teams': ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']}
print('All the keys in dictionary :',dict1.keys())
print('\nAll the Values in dictionary :',dict1.values())
print('\nAll the Items in dictionary :',dict1.items())
```
