## Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.


**Seaborn**

In Todayâ€™s World, There Is A Large Amount Of Data Is Present In Structured And Unstructured Form And To Understand This Data By Reading Is Very Very Difficult The Best Way To Understand This Data Is To Convert It Into Visualization Form To Do This Seaborn Is One Of The Visualization Libraries In Python, Which Helps To Draw Statistical Graphics With A High-Level Interface. ### Type Of Charts/Graphs/Plots :

* Line Plot

**Line Plot**

A Line Plot Is The Simplest Plot In All Plotting Types, As It Is The Visualization Of A Single Function. This Plot Helps Us To See The Relationship Between X-Axis, Y-Axis And It Also Takes Some Parameters Such As Hue, Size, Color, Etc. 

**Count Plot:**

A Count Plot Is Used To Show The Counts Of Observations In Each Categorical Bin Using Bars.

This Method Is Accepting The Parameters X, Y  Which Take The Name Of A Variable In Data, Hue It Is An Optional Parameter It Helps To Take Column Name For Color Encoding. Data Is Also An Optional Parameter That Takes DataFrame, Array, Or List Of Arrays, Dataset For Plotting.

**Bar chart**

A Bar Chart Is A Way Of Comparing A Set Of Categorical Data. It Is Better To Convert Continuous Data To Bins Before Plotting. The Bar Chart Displays Data

Using Several Bars, Each Representing A Particular Category. This Method Is Accepting The Parameters X, Y  Which Take The Name Of A Variable In Data, Hue It Is An Optional Parameter It Helps To Take Column Name For Color Encoding.

**Pair Plot**

Pair Plot Creates A Grid Of Axis Such That Each Numeric Variable In Data Will Create A Plot Between Each Other The Y-Axis Across A Single Row And The X-Axis Across A Single Column. The Diagonal Plots Are A Univariate Distribution Plot That Helps To Draw The Marginal Distribution Of The Data In Each Column.

A Pair Plot Pairwise Relationships With Other Columns In The Data Frame And Also Plot Pair Plot With Itself.

**Scatter Plot**

Scatter Plot Is The Same As A Line Plot, In A Line Plot Instead Of Points Being Joined By Line Segments, The Points Are Shown Individually With A Dot, Circle, Or Any Other Shape. The Position Of Each Marker On The Horizontal And Vertical Axis Indicates Values For An Individual Data Point. This Plot Is Used To Observe Relationships Between Variables.

This Method Is Accepting The Parameters X, Y  Which Take The Name Of A Variable In Data, Hue It Is An Optional Parameter It Helps To Take Column Name For Color Encoding. Marker Is A Parameter That Is Used To Change The Shape Of The Point 

## Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
### Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
# Loading fmri dataset using seaborn library

import seaborn as sns 
import matplotlib.pyplot as plt

#This is used to display the graph inside the notebook
%matplotlib inline  


df = sns.load_dataset('fmri')

display(df.head()) # To display the head of the fmri dataframe
```

```python
# Display the line plot of time point and signal with respect to event

plt.figure(figsize=(5,5)) # This is used to define the figure size

sns.lineplot(x = "timepoint" , y = "signal",hue='event',data=df)
```

```python
# Display the line plot of time point and signal with respect to region

plt.figure(figsize=(5,5)) # This is used to define the figure size

sns.lineplot(x = "timepoint" , y = "signal",hue='region',data=df)
```

## Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.

```python
# Loading titanic dataset using seaborn library

import seaborn as sns 
import matplotlib.pyplot as plt

#This is used to display the graph inside the notebook
%matplotlib inline  


df = sns.load_dataset('titanic')

display(df.head()) # To display the head of the dataframe
```

```python
# Display second box plot for fare with respect to pclass

sns.set_style("whitegrid")
sns.boxplot(x = 'pclass', y = 'fare' , data = df)
```

## Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
# Loading iris dataset using seaborn library

import seaborn as sns 
import matplotlib.pyplot as plt

#This is used to display the graph inside the notebook
%matplotlib inline  


df = sns.load_dataset('flights')

df = df.pivot('month','year','passengers') # tHIS IS USED TO CHANGE THE AXIUS OF MONTH AND YEAR TO PLOT THE NUMBER VALUE IN HEATMAP

display(df.head()) # To display the head of the dataframe
```

```python
sns.heatmap(data = df)
```

```python
# TO DISPLAY THE HEATMAP IN MORE CLEAN COLOUR TO VISUALIZE EASILY I AM USING THE CMAP ATTRIBUTE

sns.heatmap(df,cmap = 'BuPu')
```

```python
# lETS VISUALIZE THE HEATMAP WITH DATAPOINTS TO VISUALIZE MORE

sns.heatmap(df,cmap = 'BuPu',annot= True,fmt='d')
```
