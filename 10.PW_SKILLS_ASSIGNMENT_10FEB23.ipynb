### Q1. Which function is used to open a file? What are the different modes of opening a file? Explain each mode of file opening.


Ans: open() function is used to open a file. The syntax of the open function is:

* **open(
    file,
    mode='r',
    buffering=-1,
    encoding=None,
    errors=None,
    newline=None,
    closefd=True,
    opener=None,
)**

Python provides an open() function that accepts two arguments, file name and access mode in which the file is accessed. The function returns a file object which can be used to perform various operations.

Different modes of opening the files and its explanation are mentiooned below:

**'r'       open for reading (default)**

**'w'       open for writing, truncating the file first**

**'x'       create a new file and open it for writing**

**'a'       open for writing, appending to the end of the file if it exists**

**'b'       binary mode**

**'r+'      we can read and write the file, but the file pointer position is at the beginning of the file; if we write the     file directly, it will overwrite the beginning content.**

**'w+'      Adds reading file to the existing w mode.**



```python
# Example code to explain the open function in write mode,'w' open for writing, truncating the file first 
import os
filename = input('Enter the file name you want to open: ') 

# This code is written to get the current working directory
path = os.path.join(os.getcwd(),filename)

try :
    file = open(path,'w+') # This code is opening the file in write mode, It will create the file if doesn't exist.
    file.write('I am a data science student')
      
except FileNotFoundError as e :
    print(e)
finally:
    file.close()
```

```python
# Example code to explain open function in read mode,'r' open for reading (default)
import os
filename = input('Enter the file name you want to open: ') 

# This code is written to get the current working directory
path = os.path.join(os.getcwd(),filename)
try :
    file = open(path,'r')  # code to open the file in the reading mode if file exist
    print(file.read())
except FileNotFoundError as e:
    print(e)
finally:
    file.close()
```

```python
# Example code to explain open function in 'x' create a new file and open it for writing
import os
filename = input('Enter the file name you want to open: ') 

# This code is written to get the current working directory
path = os.path.join(os.getcwd(),filename)
try :
    file = open(path,'x')  
    print(file.write('I am currently Enrolled in PW_SKILLS masters class'))
except FileNotFoundError as e:
    print(e)
finally:
    file.close()
```

```python
# Example code to explain open function in 'a' open for writing, appending to the end of the file if it exists
import os
filename = input('Enter the file name you want to open: ') 

# This code is written to get the current working directory
path = os.path.join(os.getcwd(),filename)
try :
    file = open(path,'a')  
    print(file.write('\nI love learning python'))
except FileNotFoundError as e:
    print(e)
finally:
    file.close()
```

```python
# Example code to explain open function in 'b' binary mode
import os
filename = input('Enter the file name you want to open: ') 

# This code is written to get the current working directory
path = os.path.join(os.getcwd(),filename)
def write_binary(path):
    try :
        file = open(path,'wb') # It will open the file in write binary mode  
        num=[5, 10, 15, 20, 25]
        arr=bytearray(num)
        file.write(arr)
    
    except FileNotFoundError as e:
        print(e)
    finally:
        file.close()
        
def read_binary(path):
    try :
        file = open(path,'rb') # It will open the file in write binary mode  
        print(file.read())
    
    except FileNotFoundError as e:
        print(e)
    finally:
        file.close()
        
write_binary(path)
read_binary(path)
        

```

```python
# Example code to explain open function in 'w+' mode that Adds reading file to the existing w mode.

import os
filename = input('Enter the file name you want to open: ') 

# This code is written to get the current working directory
path = os.path.join(os.getcwd(),filename)

f = open(path, 'w+')   # create a new file or truncates it
f.write("test 1\n")
f.write("test 2\n")
f.write("test 3\n")             # now the file pointer is at the end
f.seek(0)                       # move the file pointer to the beginning
lines = f.read()                # read it, now we can read!
print(lines) 
f.close()
```

```python
# Example code to explain open function in 'r+' mode that Adds reading file to the existing w mode.

import os
filename = input('Enter the file name you want to open: ') 

# This code is written to get the current working directory
path = os.path.join(os.getcwd(),filename)

# alternative, open with `a` mode
f = open(path, 'r+')
f.seek(0)
print(f.read())                # move file position to the end of the file.
f.write("new line \n")
f.close()
```

### Q2. Why close() function is used? Why is it important to close a file?


**Why close()** :  Once all the operations are done on the file, we must close it through our Python script using the close() method. Any unwritten information gets destroyed once the close() method is called on a file object.

**Importance of close()** : 

* It releases the resources that have been tied up with the file. By this space in the RAM can be better utilized and ensures a better performance.

* It ensures better garbage collection.

* There is a limit to the number of open files in an application. It is always better to close the file to ensure that the limit is not crossed.

* If you open the file in write or read-write mode, you don’t know when data is flushed.

### Q3. Write a python program to create a text file. Write ‘I want to become a Data Scientist’ in that file. Then close the file. Open this file and read the content of the file.


```python
import os
filename = input('Enter the file name you want to open: ') 

# This code is written to get the current working directory
path = os.path.join(os.getcwd(),filename)
def write_file(path):
    try :
        file = open(path,'w') # It will open the file in write mode  
        file.write('I want to become a Data Scientist')
    
    except FileNotFoundError as e:
        print(e)
    finally:
        file.close()
        
def read_file(path):
    try :
        file = open(path,'r') # It will open the file in read mode  
        print(file.read())
    
    except FileNotFoundError as e:
        print(e)
    finally:
        file.close()
        
write_file(path) # Calling a function to write in the file
read_file(path) # Calling the function to read the file
        
```

### Q4. Explain the following with python code: read(), readline() and readlines().


 **read()** : Reads and returns a string of n characters, or the entire file as a single string if n is not provided.


**readline()** : Returns the next line of the file with all text up to and including the newline character. If n is provided as a parameter than only n characters will be returned if the line is longer than n.



**readlines()**:Returns a list of strings, each representing a single line of the file. If n is not provided then all lines of the file are returned. If n is provided then n characters are read but n is rounded up so that an entire line is returned.



```python
import os
filename = input('Enter the file name you want to open: ') 

# This code is written to get the current working directory
path = os.path.join(os.getcwd(),filename)

# This code is written to get the current working directory
path = os.path.join(os.getcwd(),filename)
def write_file(path):
    try :
        file = open(path,'w') # It will open the file in write mode  
        file.write('I want to become a Data Scientist')
        file.write('\nI learn Python Coding')
        file.write('\nMentors are Sudhanshu Kumar and Krish Naik')
    except FileNotFoundError as e:
        print(e)
    finally:
        file.close()
        
def read_file(path):
    try :
        file = open(path,'r') # It will open the file in read mode 
        print('==========Output from read()==================================')
        print(file.read()) 
        file.seek(0)
        print('==========Output from readline()==============================')
        print(file.readline())
        file.seek(0)
        print('==========Output from readlines()==============================')
        print(file.readlines())
    
    except FileNotFoundError as e:
        print(e)
    finally:
        file.close()
        
write_file(path) # Calling a function to write in the file
read_file(path) # Calling the function to read the file
```

### Q6. Explain the write() and writelines() functions. Give a suitable example.

**write()**: Use to write a string into a file. It only accepts a string as the argument. Else it will raise a TypeError: write() argument must be str.

**writelines()**	Use to write a list of strings into a file. It accepts both string and list as the argument.

```python
import os
filename1 = input('Enter the file name you want to write string: ')
filename2 = input('Enter the file name you want write lines: ')

path = os.path.join(os.getcwd(),filename1) # Creating two file for write() and for writelines()
path1 = os.path.join(os.getcwd(),filename2)

def write_file_write(path):    
    try :
        with open(path,'w') as file: # It will open the file in write mode  
            file.write('I want to become a Data Scientist')

    except FileNotFoundError as e:
        print(e)
    finally:
        file.close()

def write_file_writelines(path):    
    try :
        with open(path,'w') as file: # It will open the file in write mode 
            lines = ['I want to become a Data Scientist','I learn Python Coding','Mentors are Sudhanshu Kumar and Krish Naik']
            file.writelines(lines)
    except FileNotFoundError as e:
        print(e)
    finally:
        file.close()
        
def read_file_write(path):
    try :
        with open(path,'r') as file: # It will open the file in read mode 
            print('==========Output from write function()==================================')
            print(file.read()) 
    
    except FileNotFoundError as e:
        print(e)
    finally:
        file.close()

def read_file_writelines(path):
    try :
        with open(path,'r') as file: # It will open the file in read mode 
            print('==========Output from writelines function()==================================')
            print(file.read()) 
           
    except FileNotFoundError as e:
        print(e)
    finally:
        file.close()
        
write_file_write(path) # Calling a function to write in the file
write_file_writelines(path1) # Calling the function to writelines
read_file_write(path)
read_file_writelines(path1)
```
