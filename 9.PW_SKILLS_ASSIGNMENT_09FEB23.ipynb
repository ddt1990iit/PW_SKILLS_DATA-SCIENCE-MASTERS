## Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance.

**Multiple Inheritance** : Multiple inheritance is a feature of some object-oriented computer programming languages in which an object or class can inherit characteristics and features from more than one parent object or the parent class .



**Syntax of Multiple Inheritance :**

* Class Base1: Body of the class

* Class Base2: Body of the class

* Class Derived(Base1, Base2): Body of the class



```python
# Code for Demonstration of Multiple Inheritance

# Parent Class
class A:
    def func1():
        print('This is from the Parent Class A')
        

class B:
    def func2():
        print('This is from the Parent Class B')
        
# Child Class
class C(A,B): # Multiple inheritance
    def func3():
        print('This is from the Child Class C')
        
child_c = C() # Object of Child Class
C.func1()
C.func2()
C.func3()
```

### Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this
### class.

**Getters:** These are the methods used in Object-Oriented Programming (OOPS) which helps to access the private attributes from a class.

**Setters:** These are the methods used in OOPS feature which helps to set the value to private attributes in a class.

```python
class example():
    def __init__(self,a):
        self.__a = a # Declaration of Private Variable
        
    # Getter method
    
    def get_a(self):
        return self.__a
    
    # Setter Method
    
    def set_a(self,a):
        self.__a = a
        

ex = example(12)

print('\nThe initial value of the private variable before setter:' ,ex.get_a()) # Getting a private variable from class example through object

ex.set_a(14) # Setting a different value to the private variable of class using setter method through a object.

print('\nThe changed value of the private variable after setter method:',ex.get_a())
    
```

### Q5.What is method overriding in python? Write a python code to demonstrate method overriding.

**Method Overriding** : It is a type of Run-time Polymorphism. A child class method overrides (or provides its implementation) the parent class method of the same name, parameters, and return type. It is used to over-write (redefine) a parent class method in the derived class.

```python
# Python code to demonstrate method overriding

class A:
    def func1(self):
        print('This is the First Function from class A')
        
    def func2(self):
        print('This is the Second Function from Class A')
        

class B(A):# Inheriting the class A
    def func1(self):
        print('This is the redefine function of A in derived Class B') # Method Overriding
        
    def display_att(self):
        print('This is the display attribute function from Class B')
        
object_B = B()
object_A = A()

object_B.func1()
object_A.func1()
    
    
```
