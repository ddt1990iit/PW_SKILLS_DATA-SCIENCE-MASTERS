```python
# Code to input the Number from the user

while True:
    print('\n============================================================================================')
    reply = input('Enter yes or no if you want to continue  ')
    reply1 = reply.upper()
    
    if reply1 == 'YES':
        count = 0  # Initialization of Count Variable 
        num_input = int(input('\nEnter the Number greater than 10 to check for divisibilty by 3 '))
        print('\nEntered Number is {}'.format(num_input))
        
        # Code to calculate the count of divisibilty of number by 3
        
        num = num_input  
        
        while num>=10:
            if num % 3 == 0:
                count = count+ 1 # Increment the count variable
                num = num -1 # Decrement the count Variable 
            else:
                num = num-1
                
        print('\nThe Number entered by you is {} and it is {} times divisible by 3'.format(num_input,count))
    
    else:
        print('\nThank You for your interest')
        break
            


```

### **QUESTION 5:**

Ans: 

**Why ??** : While loop is used to execute the block of the code until a given condition is satisfied.It is used because while loop falls under the category of indefinite iteration.Indefinite itertion means the number of times loop is executed is not specified in advance.

**When ??** : It is used when we have to run a block of code several times using specified condition and it checks the condition during the runtime that how many time the code has to be executed. It is used to do iteration.


```python
# Example 1:

# Addition of even number upto certain length:

num1 = int(input('Enter the number upto which you want to calculate the sum of even numbers '))
sum = 0
# Code to perform the Indefinite iteration

while num1 > 0:
    if num1 % 2 == 0:
        sum = sum +num1
    num1 = num1 -1   # This decrement will decide for how much time code will be executed
    
print('\nSum of even numbers upto the given number is {}'.format(sum))


```

```python
# Example 2 : Calculate the factorial of the number

number = int(input('Enter the number to calulate the factorial: '))
number1 = number # Temporary Assignment
fact = 1
if number1 == 0:
    fact =1
else:
    while number1 > 0:
        fact = fact * number1
        number1 = number1 - 1
print('Factorial of number {} is {}'.format(number,fact))
```

```python
# Code to print the pattern shown image above using while loop
n = int(input('Enter the number of rows to print in Pyramid'))
count = 1
num = 1
temp = 1

# Loop to check for the rows
while num <= n:
    temp = num  # Assignment of temp variable for reduced spaces
    
    # Loop to print the spaces in decrement order
    
    while temp < n:
        print(' ',end='')
        temp = temp+1
        
    # Loop to print *  
    
    while count <= num:
        print('*',end=' ')
        count = count + 1
    print('\r')
    count = 1
    num = num + 1
```

**Pattern 3**


```python
n = int(input('Enter the number of rows for Pyramid'))
num = 1
temp = 1
# Code to iterate through rows
while temp <= n:
    # Code to print * sequentially
    while num <= temp:
        print('*',end=' ')
        num = num+1
    print('\r')
    num = 1 
    temp = temp+1
```
